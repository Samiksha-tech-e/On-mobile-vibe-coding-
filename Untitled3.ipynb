```python

import cv2
import numpy as np
import base64
from IPython.display import display, Javascript
from google.colab.output import eval_js

def video_stream():
    js = Javascript('''
    var video;
    var div = null;
    var stream;
    var captureCanvas;

    function pendingCreate() {
      if (div != null) {
        removeDom();
      }
      div = document.createElement('div');
      div.style.border = '2px solid black';
      div.style.padding = '10px';
      div.style.width = '100%';
      div.style.maxWidth = '600px';
      document.body.appendChild(div);

      video = document.createElement('video');
      video.style.display = 'block';
      video.width = 320;
      video.height = 240;
      video.setAttribute('playsinline', '');

      var constraints = {
        video: {
          facingMode: "user",
          width: { ideal: 320 },
          height: { ideal: 240 }
        }
      };

      navigator.mediaDevices.getUserMedia(constraints).then(function(str) {
        stream = str;
        video.srcObject = stream;
        video.play();
      });

      div.appendChild(video);
    }

    function removeDom() {
      if (stream) stream.getVideoTracks()[0].stop();
      if (video) video.remove();
      if (div) div.remove();
      video = null;
      div = null;
    }

    async function takePhoto(quality) {
      // CRITICAL CHECK: Don't send data if video stream isn't fully playing yet
      if (video == null || video.videoWidth === 0 || video.readyState < 2) {
        return "WAIT";
      }
      if (captureCanvas == null) {
        captureCanvas = document.createElement('canvas');
        captureCanvas.width = video.videoWidth;
        captureCanvas.height = video.videoHeight;
      }
      captureCanvas.getContext('2d').drawImage(video, 0, 0);
      return captureCanvas.toDataURL('image/jpeg', quality);
    }

    pendingCreate();
    ''')
    display(js)

def js_to_image(js_reply):
    # Safety checks to catch blank or initializing signals
    if not js_reply or js_reply == "WAIT" or "," not in js_reply:
        return None
    try:
        image_bytes = base64.b64decode(js_reply.split(',')[1])
        jpg_as_np = np.frombuffer(image_bytes, dtype=np.uint8)
        img = cv2.imdecode(jpg_as_np, flags=1)
        return img
    except Exception:
        return None
```

```python

import mediapipe as mp
import cv2
import base64
from google.colab.output import eval_js
from IPython.display import display, Javascript

# Explicit Tasks Framework Mapping
BaseOptions = mp.tasks.BaseOptions
HandLandmarker = mp.tasks.vision.HandLandmarker
HandLandmarkerOptions = mp.tasks.vision.HandLandmarkerOptions
VisionRunningMode = mp.tasks.vision.RunningMode

HAND_PALM_CONNECTIONS = [(0, 1), (1, 2), (2, 3), (3, 4), (0, 5), (5, 9), (9, 13), (13, 17), (0, 17)]
HAND_FINGER_CONNECTIONS = [
    (5, 6), (6, 7), (7, 8),      # Index
    (9, 10), (10, 11), (11, 12),  # Middle
    (13, 14), (14, 15), (15, 16), # Ring
    (17, 18), (18, 19), (19, 20)  # Pinky
]
ALL_CONNECTIONS = HAND_PALM_CONNECTIONS + HAND_FINGER_CONNECTIONS

# Initialize stream engine
video_stream()

display(Javascript('''
  var img = document.getElementById("processed_img");
  if (!img) {
    img = document.createElement('img');
    img.id = 'processed_img';
    document.body.appendChild(img);
  }
'''))

options = HandLandmarkerOptions(
    base_options=BaseOptions(model_asset_path='hand_landmarker.task'),
    running_mode=VisionRunningMode.IMAGE,
    num_hands=1
)

with HandLandmarker.create_from_options(options) as landmarker:
    while True:
        js_reply = eval_js('takePhoto(0.6)')
        if not js_reply or js_reply == "WAIT":
            continue

        frame = js_to_image(js_reply)
        if frame is None:
            continue

        # --- FIX 1: UN-MIRROR THE VIDEO ---
        # cv2.flip(frame, 1) flips the video horizontally to fix mirroring issues
        frame = cv2.flip(frame, 1)

        image_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        mp_image = mp.Image(image_format=mp.ImageFormat.SRGB, data=image_rgb)

        detection_result = landmarker.detect(mp_image)

        gesture_text = "Searching Hand..."

        if detection_result.hand_landmarks:
            h, w, _ = frame.shape
            for hand_landmarks in detection_result.hand_landmarks:

                # Draw connections (Green skeleton)
                for connection in ALL_CONNECTIONS:
                    start_point = (int(hand_landmarks[connection[0]].x * w), int(hand_landmarks[connection[0]].y * h))
                    end_point = (int(hand_landmarks[connection[1]].x * w), int(hand_landmarks[connection[1]].y * h))
                    cv2.line(frame, start_point, end_point, (0, 255, 0), 2)

                # Draw tracking joints (Blue dots)
                for landmark in hand_landmarks:
                    cx, cy = int(landmark.x * w), int(landmark.y * h)
                    cv2.circle(frame, (cx, cy), 4, (255, 0, 0), -1)

                # --- FIX 2 & 3: GESTURE DETECTION ALGORITHM ---
                # Check if finger tips are higher up on the screen than their lower joints.
                # Note: In computer vision, the top of the screen is y = 0, so "higher up" means a SMALLER y value.

                index_open  = hand_landmarks[8].y  < hand_landmarks[6].y
                middle_open = hand_landmarks[12].y < hand_landmarks[10].y
                ring_open   = hand_landmarks[16].y < hand_landmarks[14].y
                pinky_open  = hand_landmarks[20].y < hand_landmarks[18].y

                # Case A: All fingers closed -> FIST
                if not index_open and not middle_open and not ring_open and not pinky_open:
                    gesture_text = "Fist Detected!"

                # Case B: Only the index finger is open -> SINGLE FINGER
                elif index_open and not middle_open and not ring_open and not pinky_open:
                    gesture_text = "Single Finger Detected!"

                # Case C: Open Palm / Other movements
                else:
                    gesture_text = "Tracking Hand..."

        # Paint the detected gesture label text onto the video frame
        cv2.putText(frame, gesture_text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2, cv2.LINE_AA)

        # Encode and send back to HTML template view container
        _, bbox_array = cv2.imencode('.jpg', frame)
        bbox_bytes = base64.b64encode(bbox_array).decode('utf-8')

        eval_js(f'document.getElementById("processed_img").src="data:image/jpeg;base64,{bbox_bytes}"')
```
